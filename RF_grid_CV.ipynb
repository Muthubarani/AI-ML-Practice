```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Social_Network_Ads.csv")
df
```

```python
df = pd.get_dummies(df, dtype= int, drop_first= True, )
df
```

```python
x = df[["Age","EstimatedSalary","Gender_Male"]]
y = df["Purchased"]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import cross_val_predict, GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100,150],
    'max_depth': [5,10],
    'min_samples_split': [2,5],
    'min_samples_leaf': [1,2]
}

grid = GridSearchCV(RandomForestClassifier(), param_grid, refit = True,verbose= 3,n_jobs= -1,cv= 5)
grid.fit(x, y)
```

```python
y_pred = cross_val_predict(grid.best_estimator_, x, y)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score
cm = confusion_matrix(y, y_pred)
clf_report = classification_report(y, y_pred)
f1_macro = f1_score(y, y_pred, average='weighted')
```

```python
print("Best Parameters:", grid.best_params_)
print("The f1_macro:", f1_macro)
print("The confusion Matrix:\n", cm)
print("The report:\n", clf_report)
```

```python
import pickle
filename = "SVC_CV_best_model.sav"
pickle.dump(grid.best_estimator_, open(filename, 'wb'))
```

```python
results_df = pd.DataFrame(grid.cv_results_)
results_df.head()
```

```python
# Get input from the user
age = int(input("Enter Age: "))
estimated_salary = int(input("Enter Estimated Salary: "))
gender_male = int(input("Enter Gender (0 for Female, 1 for Male): "))

# Create a DataFrame from user input
user_input = pd.DataFrame([[age, estimated_salary, gender_male]], columns=["Age", "EstimatedSalary", "Gender_Male"])

# Preprocess the user input using the same scaler
user_input_scaled = sc.transform(user_input)

# Load the saved model
loaded_model = pickle.load(open(filename, 'rb'))

# Make a prediction
prediction = loaded_model.predict(user_input_scaled)

# Print the prediction
if prediction[0] == 1:
    print("The model predicts that the user will purchase.")
else:
    print("The model predicts that the user will not purchase.")
```
